```python
import torch
import numpy as np
```

```python
tensor_a = torch.tensor([1,2,3,4])
tensor_b = torch.tensor([5,6,7,8])
```

```python
#Addition
tensor_a + tensor_b
```

```python
# Addition Longhand i.e function
torch.add(tensor_a,tensor_b)
```

```python
 #subtraction
tensor_b - tensor_a
```

```python
torch.sub(tensor_b,tensor_a)
```

```python
#multiplication
tensor_a * tensor_b
```

```python
torch.mul(tensor_a,tensor_b)
```

```python
#Division
tensor_b / tensor_a
```

```python
torch.div(tensor_b,tensor_a)
```

```python
# Remainder Modulus
tensor_b % tensor_a
```

```python
torch.remainder(tensor_b,tensor_a)
```

```python
#Exponents / Power
torch.pow(tensor_a,tensor_b)
```

```python
# Reassignment
tensor_a = tensor_a + tensor_b
```
